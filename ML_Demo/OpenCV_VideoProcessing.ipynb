```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

writer=cv2.VideoWriter('MyFirstVideo.mp4', cv2.VideoWriter_fourcc(*'DIVX'), 20, (width, height))

while True:
    ret, frame = cap.read()
    
    writer.write(frame)
    
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        

cap.release()
writer.release()
cv2.destroyAllWindows()
```

```python
cap2 = cv2.VideoCapture('MyFirstVideo.mp4')

if cap2.isOpened() == False:
    print('ERROR FILE NOT FOUND OR WRONG CODEC USED!')
    

import time

while cap2.isOpened():
    ret, frame = cap2.read()
    
    if ret == True:
        time.sleep(1/20)  #1/frame_rate to slow down the video
        cv2.imshow('frame', frame)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    
    
    else:
        break

        
cap2.release()
cv2.destroyAllWindows();
```

```python
cap3 = cv2.VideoCapture(0)
width3 = int(cap3.get(cv2.CAP_PROP_FRAME_WIDTH))
height3 = int(cap3.get(cv2.CAP_PROP_FRAME_HEIGHT))

#Top Left Corner
x = width3 // 2   #to get integers use '//'
y = height3 // 2

#Width and Height of rectangle
w = width3 // 4
h = height3 // 4

#Bottom Right Corner x+w, y+h

while True:
    ret, frame = cap3.read()
    
    cv2.rectangle(frame, (x, y), (x+w, y+h), color=(0,0,255), thickness = 4)
    
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
        
        
        
cap3.release()
cv2.destroyAllWindows();
```

```python
#Callback function
def draw_rectangle(event, x, y, flags, param):
    global pt1, pt2, topLeft_clicked, botRight_clicked
    
    if event == cv2.EVENT_LBUTTONDOWN:
        if topLeft_clicked and botRight_clicked:
            pt1 = (0,0)
            pt2 = (0,0)
            topLeft_clicked = False
            botRight_clicked = False

        if topLeft_clicked == False:
            pt1 = (x,y)
            topLeft_clicked = True

        elif botRight_clicked == False:
            pt2 = (x,y)
            botRight_clicked = True


#Global Variables
pt1 = (0,0)
pt2 = (0,0)
topLeft_clicked = False
botRight_clicked = False


#Connect to the callback
cap4 = cv2.VideoCapture(0)
cv2.namedWindow('Test')
cv2.setMouseCallback('Test', draw_rectangle)

while True:
    ret, frame = cap4.read()
    
    #Draw the rectangle
    if topLeft_clicked:
        cv2.circle(frame, center=pt1, radius=5, color=(0,0,255), thickness=-1)
    
    if topLeft_clicked and botRight_clicked:
        cv2.rectangle(frame, pt1, pt2, (0,0,255), 5)
    
    cv2.imshow('Test', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
        
        
cap4.release()
cv2.destroyAllWindows();
```

```python
import cv2

#Callback function
def draw_circle(event, x, y, flags, param):
    global center, clicked_up
    
    if event == cv2.EVENT_LBUTTONDOWN:
        center = (x,y)
        clicked_up = False
    
    if event == cv2.EVENT_LBUTTONUP:
        clicked_up = True


#Global Variables
center = (0,0)
clicked_up = False


#Connect to the callback function
cap5 = cv2.VideoCapture(0)
cv2.namedWindow('Test')
cv2.setMouseCallback('Test', draw_circle)

while True:
    ret, frame = cap5.read()
    
    #Draw the rectangle
    if clicked_up:
        cv2.circle(frame, center, radius=20, color=(255,0,0), thickness=2)
    
    cv2.imshow('Test', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
  
        
cap5.release()
cv2.destroyAllWindows();
```
